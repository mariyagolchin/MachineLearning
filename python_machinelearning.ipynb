```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x= np.arange(1,10)
y= np.array([33, 55, 66, 43, 50, 52, 60,70, 89])
# plt.scatter(x,y)
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.yticks(y,[33, 55, 66, 43, 50, 52, 60,70, 89])
plt.show()
```

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
red=LinearRegression()
red.fit(x,y)
y_=  red.predict(x)

x= np.arange(1,10)
y= np.array([33, 55, 66, 43, 50, 52, 60,70, 89])
# plt.scatter(x,y)
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x,y_)
plt.yticks(y,[33, 55, 66, 43, 50, 52, 60,70, 89])
plt.show()
```
